```python
import pandas as pd
import datetime as dt
import numpy as np


df = pd.read_csv('Train.csv')

df.head()


```

```python
df.shape
```

```python

df["STATUS"].unique()

```

```python

df.isnull().values.any()
```

```python

df.head()
```

```python


df =df.drop('STATUS', axis=1)
df =df.drop('DATOP', axis=1)
df['STD_month'] = df['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").month)
df['STD_DAY'] = df['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").day)
df['STD_houre'] = df['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").hour)
df['STD_MINUTE'] = df['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").minute)


df['STA_month'] = df['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").month)
df['STA_DAY'] = df['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").day)
df['STA_houre'] = df['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").hour)
df['STA_MINUTE'] = df['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").minute)

df =df.drop('STD', axis=1)
df =df.drop('STA', axis=1)
df.head()

```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression() 

y_train=df['target']
X_train=df.drop('target', axis=1)
X_train=X_train.drop('ID', axis=1)
X_train=X_train.drop('FLTID', axis=1)

X_train.head()


 
```

```python
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
X_train['AC']= label_encoder.fit_transform(X_train['AC']) 
X_train['DEPSTN']= label_encoder.fit_transform(X_train['DEPSTN']) 
X_train['ARRSTN']= label_encoder.fit_transform(X_train['ARRSTN']) 
```

```python
X_train.head()
```

```python
regressor.fit(X_train, y_train) 
```

```python
d_test = pd.read_csv('Test.csv')
ids=d_test['ID'].values
```

```python
d_test.head()
```

```python
d_test=d_test.drop('STATUS', axis=1)
d_test=d_test.drop('DATOP', axis=1)
d_test['STD_month'] = d_test['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").month)
d_test['STD_DAY'] = d_test['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").day)
d_test['STD_houre'] = d_test['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").hour)
d_test['STD_MINUTE'] = d_test['STD'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").minute)


d_test['STA_month'] = d_test['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").month)
d_test['STA_DAY'] = d_test['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").day)
d_test['STA_houre'] = d_test['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").hour)
d_test['STA_MINUTE'] = d_test['STA'].apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d %H.%M.%S").minute)

d_test =d_test.drop('STD', axis=1)
d_test =d_test.drop('STA', axis=1)
d_test=d_test.drop('ID', axis=1)
d_test=d_test.drop('FLTID', axis=1)
```

```python
label_encoderrr = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
d_test['AC']= label_encoderrr.fit_transform(d_test['AC']) 
d_test['DEPSTN']= label_encoderrr.fit_transform(d_test['DEPSTN']) 
d_test['ARRSTN']= label_encoderrr.fit_transform(d_test['ARRSTN']) 
```

```python
d_test.head()
```

```python
y_pred = regressor.predict(d_test) 
```

```python
y_pred=np.round_(y_pred,3)
result=np.column_stack((ids,y_pred))
```

```python
print(result)
```

```python
headers = ['id','target']
df7 = pd.DataFrame(result, columns=headers)
df7.to_csv('result1.csv' , index=False)
```
