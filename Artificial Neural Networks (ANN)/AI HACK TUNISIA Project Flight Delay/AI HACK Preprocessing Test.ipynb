## Extract data from csv

```python

raw_csv_data = pd.read_csv('Test.csv')

unscaled_inputs_all = raw_csv_data.drop(['FLTID','AC'], axis=1)
```

## Transform dates

```python
from datetime import datetime

unscaled_inputs_all['DATOP'] =  pd.to_datetime(unscaled_inputs_all['DATOP'])
unscaled_inputs_all['STD'] =  pd.to_datetime(unscaled_inputs_all['STD'])
unscaled_inputs_all['STA'] =  pd.to_datetime(unscaled_inputs_all['STA'],format = '%Y-%m-%d %H.%M.%S')
```

```python
unscaled_inputs_all['DATOP_Year'] =  unscaled_inputs_all['DATOP'].map(lambda x: x.year)
unscaled_inputs_all['DATOP_Month'] =  unscaled_inputs_all['DATOP'].map(lambda x: x.month)
unscaled_inputs_all['DATOP_Day'] =  unscaled_inputs_all['DATOP'].map(lambda x: x.day)
unscaled_inputs_all['DATOP_Weekday'] =  unscaled_inputs_all['DATOP'].map(lambda x: x.weekday())
unscaled_inputs_all['DATOP_Season'] =  unscaled_inputs_all['DATOP'].map(lambda x: (x.month%12 + 3)//3)
```

```python
unscaled_inputs_all['STD_Year'] =  unscaled_inputs_all['STD'].map(lambda x: x.year)
unscaled_inputs_all['STD_Month'] =  unscaled_inputs_all['STD'].map(lambda x: x.month)
unscaled_inputs_all['STD_Day'] =  unscaled_inputs_all['STD'].map(lambda x: x.day)
unscaled_inputs_all['STD_Hour'] =  unscaled_inputs_all['STD'].map(lambda x: x.hour)
unscaled_inputs_all['STD_Minute'] =  unscaled_inputs_all['STD'].map(lambda x: x.minute)
unscaled_inputs_all['STD_Weekday'] =  unscaled_inputs_all['STD'].map(lambda x: x.weekday())
unscaled_inputs_all['STD_Season'] =  unscaled_inputs_all['STD'].map(lambda x: (x.month%12 + 3)//3)
```

```python
unscaled_inputs_all['STA_Year'] =  unscaled_inputs_all['STA'].map(lambda x: x.year)
unscaled_inputs_all['STA_Month'] =  unscaled_inputs_all['STA'].map(lambda x: x.month)
unscaled_inputs_all['STA_Day'] =  unscaled_inputs_all['STA'].map(lambda x: x.day)
unscaled_inputs_all['STA_Hour'] =  unscaled_inputs_all['STA'].map(lambda x: x.hour)
unscaled_inputs_all['STA_Minute'] =  unscaled_inputs_all['STA'].map(lambda x: x.minute)
unscaled_inputs_all['STA_Weekday'] =  unscaled_inputs_all['STA'].map(lambda x: x.weekday())
unscaled_inputs_all['STA_Season'] =  unscaled_inputs_all['STA'].map(lambda x: (x.month%12 + 3)//3)
```

```python
unscaled_inputs_all = unscaled_inputs_all.drop(['DATOP','STD','STA'], axis=1)
```
